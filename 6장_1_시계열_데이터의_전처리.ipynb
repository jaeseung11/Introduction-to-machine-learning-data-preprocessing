```python
import pandas as pd

# 파일 리드
dat_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/머신러닝 데이터 전처리 입문_로드북/chap6/data/energydata.csv', sep=',') 

# 선두에서 5행까지 표시
dat_df.head(10)

# 어떤 가정의 가전제품과 조명 등의 상태 로그
```

```python
# 데이터의 행수, 열수를 확인
print(dat_df.shape)
```

```python
# 데이터 형의 확인
print(dat_df.dtypes)
```

```python
# date를 object형에서 datetime형으로 변환 # 왜 데이터셋에서 Timestamp로 되어 있는데 또 포맷을 쓴것은 그냥 변환이 안되서
dat_df['date'] = pd.to_datetime(dat_df['date'], format='%Y-%m-%d %H:%M:%S')

# date열의 데이터형을 확인
print(dat_df['date'].dtypes)
# date 첫번째 데이터의 형을 확인
print(type(dat_df['date'][0]))
```

```python
dat_df['date']
```

```python
# 전행과 시간차를 계산
dat_df['dif_min'] = dat_df['date'].diff().dt.total_seconds()/60
# 선두요소의 결손을 0으로 메꾼다
dat_df['dif_min'] = dat_df['dif_min'].fillna(0)
# 선두에서 5행까지 표시
dat_df['dif_min'].head()
```

```python
# 경과시간(분)의 계산
dat_df['cum_min'] = dat_df['dif_min'].cumsum()

# 원래의 날짜와 경과시간(분)을 표시
dat_df[['date', 'cum_min']].head()
```

```python
# 연습문제1
# 경과시간(시간)의 계산
dat_df['cum_hour'] = (dat_df['cum_min']/60).round(2).astype(int)

# 1시간 단위의 연결번호를 확인
print(dat_df['cum_hour'].unique())

# 원래의 날짜와 경과시간(시)을 표시元
dat_df[['date', 'cum_min', 'cum_hour']].head(10)
```

#### 결손값의 확인

```python
# 데이터 행별 결손값의 개수
print(dat_df.isnull().sum(axis=1).sort_values(ascending=False))

# 데이터 열별 결손값의 개수
print(dat_df.isnull().sum(axis=0))
```

#### 통계량의 계산

```python
# 데이터 형이 수치인 항목의 각종 툥계량을 계산
dat_df.describe()
```

```python
# 연습문제2
# 긱힝목의 평균값
dat_df.groupby('cum_hour').mean()
```

```python
# 각 항목의 표준편차
dat_df.groupby('cum_hour').std()
```

#### 데이터의 가시화

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Appliances의 꺽은선 그래프
plt.plot(dat_df['cum_min'], dat_df['Appliances'])

# y축(세로축)의 레이블을 추가
plt.xlabel('cum_min')
plt.ylabel('Appliances')
plt.show()
```

```python
# 연습문제3
# T1~T3의 꺽으선 그래프
plt.plot(dat_df['cum_min'], dat_df['T1'], '-r', label='T1')
plt.plot(dat_df['cum_min'], dat_df['T2'], '-g', label='T2')
plt.plot(dat_df['cum_min'], dat_df['T3'], '-b', label='T3')

# 레이블을 추가
plt.xlabel('cum_min')
plt.legend()
plt.show()
```
